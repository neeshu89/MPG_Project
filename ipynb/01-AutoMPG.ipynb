# Auto - MPG Data Set

```r
cars = read.table('http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header = F)
```

## Size of data set

```r
dim(cars)
head(cars)
```

## Memory of data set

```r
object.size(cars)
```

## Data Types

```r
colnames(cars) <- c('mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name')
```

```r
head(cars)
```

```r
lapply(cars, class)
```

## Summary Statistics
Ignore Horsepower and Car Name as they are stored as factors

```r
summary(cars)
```

```r
sapply(cars, function(x) mean(as.numeric(x), na.rm = TRUE))
```

```r
sapply(cars, function(x) median(as.numeric(x), na.rm = TRUE))
```

```r
sapply(cars, function(x) sd(as.numeric(x), na.rm = TRUE))
```

```r
sapply(cars, function(x) names(sort(-table(x)))[1])
```

## Distribution of the target

```r
library(repr)
options(repr.plot.width=12, repr.plot.height=6)
hist(unlist(cars$mpg), freq = FALSE)
lines(density(cars$mpg))
```

## Distribution of other variables

```r
library(repr)
options(repr.plot.width=12, repr.plot.height=12)
pairs(cars)
```

```r
options(repr.plot.width=12, repr.plot.height=6)
hist(unlist(cars$weight), freq = FALSE)
lines(density(cars$weight))
```

```r
options(repr.plot.width=12, repr.plot.height=6)
hist(cars$'model year', freq = FALSE)
lines(density(cars$'model year'))
```

```r
options(repr.plot.width=12, repr.plot.height=6)
hist(unlist(cars$origin), freq = FALSE)
lines(density(cars$origin))
```

```r
options(repr.plot.width=12, repr.plot.height=6)
hist(unlist(cars$displacement), freq = FALSE)
lines(density(cars$displacement))
```
